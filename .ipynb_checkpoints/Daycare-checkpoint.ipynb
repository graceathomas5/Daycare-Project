```python
#map each borough (in health codes) to each county (in deserts)

#dictionary borough to counties
borough_to_county = {"MANHATTAN": "New York",
    "BROOKLYN": "Kings",
    "QUEENS": "Queens",
    "BRONX": "Bronx",
    "STATEN ISLAND": "Richmond"}

#now map every county to a borough. So county is now added to the end of this table
HealthCode["County"] = HealthCode["Borough"].map(borough_to_county)
HealthCode[["Borough", "County"]].sample(10)
```

```python
print(HealthCode.columns)
```

```python
#create new code for census tract in HealthCode
HealthCode["tract_code"] = (
    HealthCode["Census Tract"]
    .astype("Int64")
    .astype(str)
    .str.zfill(6))
HealthCode["tract_code"] 
```

```python
#convert these tract_code values to stings
HealthCode["tract_code"] = (HealthCode["Census Tract"].astype("string"))

#manually add census tracts for blanks using https://censusreporter.org/locate/
manual_tracts = {"100A ALDRICH STREET": "046205",
    "16-01 150 STREET": "001029",
    "2351 ADAM CLAYTON POWELL JR BOULEVARD": "000228",
    "472 ATLANTIC AVENUE": "000039",
    "2750 FREDERICK DOUGLASS BOULEVARD": "000234",
    "2600 ADAM CLAYTON POWELL JR BOULEVARD": "000236"}

HealthCode["tract_code"] = HealthCode["tract_code"].fillna(
    HealthCode["Street Address"].map(manual_tracts))

#verify
HealthCode[HealthCode["Street Address"].isin(manual_tracts.keys())][
    ["Street Address", "Census Tract", "tract_code"]
]
```

```python
#Do I have duplicate addresses for the same preschool?
duplicates = HealthCode[
    HealthCode.duplicated(subset=["Street Address"], keep=False)]
duplicates = duplicates.sort_values(by="Street Address")
duplicates
```

```python
import pandas as pd
```

```python
#importing violation dataset
Violations = pd.read_csv("DOHMH_Childcare_Center_Inspections_(Historical)_20260511.csv")
Violations
```

```python
#also, only allow childcare: ages 0-5
allowed_ages = ["0 YEARS - 2 YEARS", "2 YEARS - 5 YEARS", "3 YEARS - 5 YEARS"]
Violations = Violations[
    Violations["Age Range"].isin(allowed_ages)]

#validate that
Violations["Age Range"].value_counts(dropna=False)
```

```python
#Make everything title case
Violations["Center Name"] = (Violations["Center Name"].astype(str).str.strip().str.title())
Violations["Legal Name"] = (Violations["Legal Name"].astype(str).str.strip().str.title())
Violations["Street"] = (Violations["Street"].astype(str).str.strip().str.title())
Violations["Building"] = Violations["Building"].astype(str).str.strip()
Violations["Street"] = Violations["Street"].astype(str).str.strip()

Violations
```

```python
# Convert Inspection Date to datetime
Violations["Inspection Date"] = pd.to_datetime(
    Violations["Inspection Date"],
    format="%m/%d/%y",
    errors="coerce")

# Convert ZipCode to float
Violations["ZipCode"] = Violations["ZipCode"].astype("Int64")
Violations
```

```python
# Create unique center-location ID

# Clean fields
Violations["Building_clean"] = Violations["Building"].astype(str).str.strip().str.upper()
Violations["Street_clean"] = (
    Violations["Street"].astype(str).str.upper().str.strip().str.replace(r"\s+", " ", regex=True))

Violations["Phone_clean"] = (Violations["Phone"].astype(str).str.replace(r"\D", "", regex=True).str.strip())

Violations["ZipCode_clean"] = Violations["ZipCode"].astype("Int64").astype(str)

#Create one physical center/site ID

Violations["Center_Site_ID"] = (Violations["Building_clean"] + " "
    + Violations["Street_clean"] + " | "
    + Violations["ZipCode_clean"] + " | "
    + Violations["Phone_clean"])

Violations
```

```python
#don't include instances of no new violations
Violations = Violations[~Violations["Regulation Summary"].str.contains(
"There were no new violations observed at the time of this inspection/visit",
        na=False)]
Violations
```

```python
#Manually checking instances where there's diff center names of the same location and phone number. If the daycare is the same, merge
multi_name_sites = (
    Violations.groupby(["Building", "Street", "Phone"])
    .agg(
        n_names=("Center Name", "nunique"),
        center_names=("Center Name", lambda x: sorted(set(x))),
        borough=("Borough", "first"),
        zipcode=("ZipCode", "first"))
    .reset_index())

# Only keep sites with multiple names
multi_name_sites = multi_name_sites[
    multi_name_sites["n_names"] > 1]
for _, row in multi_name_sites.head(100).iterrows():
    print(f"\nAddress: {row['Building']} {row['Street']}")
    print(f"Phone: {row['Phone']}")
    print("Center Names:")
    for name in row["center_names"]:
        print(f"  - {name}")

#all instances are the same
```

```python
""" Deduplicate same violation at same physical site/date/code/summary:
So I only want to count a violation once if it happened at the same site, same date, same regulation and health code
"""

Violations_deduped = Violations.drop_duplicates(
    subset=[
        "Center_Site_ID",
        "Inspection Date",
        "Regulation Summary",
        "Health Code Sub Section"]).copy()

# Count how many times each unique violation appeared at each physical site
code_counts = (
    Violations_deduped
    .groupby([
        "Center_Site_ID",
        "Health Code Sub Section",
        "Regulation Summary"])
    .size()
    .reset_index(name="times_seen"))

# Repeat instances beyond first occurrence
code_counts["recurrence_score"] = (
    code_counts["times_seen"] - 1 ).clip(lower=0)

# Did this violation type recur?
code_counts["did_repeat"] = (
    code_counts["times_seen"] >= 2)


#Create a lookup from the deduped data for violation category
category_lookup = (
    Violations_deduped
    .assign(
        Violation_Category_Clean=(
            Violations_deduped["Violation Category"].fillna("BLANK").str.strip().str.upper()))
    .drop_duplicates(

        subset=[
            "Center_Site_ID",
            "Health Code Sub Section",
            "Regulation Summary"])
        
    [[      "Center_Site_ID",
            "Health Code Sub Section",
            "Regulation Summary",
            "Violation_Category_Clean"
        ]])

code_counts = code_counts.merge(
    category_lookup,
    on=[
        "Center_Site_ID",
        "Health Code Sub Section",
        "Regulation Summary"
    ],
    how="left")

# Count ALL deduped violations per center
total_violations = (
    Violations_deduped
    .groupby("Center_Site_ID")
    .size()
    .reset_index(name="total_violations"))

# Find the most repeated violation per center
top_repeat_violation = (
    code_counts
    .sort_values(
        ["Center_Site_ID", "times_seen"],
        ascending=[True, False])
    .drop_duplicates("Center_Site_ID")
    [ [
            "Center_Site_ID",
            "Regulation Summary",
            "times_seen"  ] ]
    .rename(columns={
        "Regulation Summary": "top_repeat_violation",
        "times_seen": "max_times_same_violation_seen"}))

# Count recurrence score by violation category. Ex: Critical issue seen 3 times = recurrence_score 2

recurrence_by_category = (
    code_counts
    .pivot_table(
        index="Center_Site_ID",
        columns="Violation_Category_Clean",
        values="recurrence_score",
        aggfunc="sum",
        fill_value=0)
    .reset_index()
)

recurrence_by_category = recurrence_by_category.rename(columns={
    "CRITICAL": "recurring_critical_violations",
    "GENERAL": "recurring_general_violations",
    "PUBLIC HEALTH HAZARD": "recurring_public_health_violations",
    "BLANK": "recurring_blank_category_violations"
})

# Summarize center-level metrics
center_recurrence = (
    code_counts
    .groupby("Center_Site_ID")
    .agg(
        recurrence_score=("recurrence_score", "sum"),
        repeated_violation_types=("did_repeat", "sum"),
        total_violation_types=("Regulation Summary", "count"))
    .reset_index())


# Count violations by category per center
violation_category_counts = (
    Violations_deduped
    .assign(
        Violation_Category_Clean=
        Violations_deduped["Violation Category"]
        .fillna("BLANK")
        .str.strip()
        .str.upper())
    .pivot_table(
        index="Center_Site_ID",
        columns="Violation_Category_Clean",
        values="Regulation Summary",
        aggfunc="count",
        fill_value=0)
    .reset_index())

# Rename columns to easier names
violation_category_counts = (
    violation_category_counts.rename(columns={
        "CRITICAL": "critical_violations",
        "GENERAL": "general_violations",
        "PUBLIC HEALTH HAZARD": "public_health_violations",
        "BLANK": "blank_category_violations"})
)

# Merge category counts into center table
center_recurrence = center_recurrence.merge(
    violation_category_counts,
    on="Center_Site_ID",
    how="left"
)

# Fill missing with 0
category_cols = [
    "critical_violations",
    "general_violations",
    "public_health_violations",
    "blank_category_violations"]

for col in category_cols:
    if col in center_recurrence.columns:
        center_recurrence[col] = (
            center_recurrence[col]
            .fillna(0)
            .astype(int))

# Merge in total violations
center_recurrence = center_recurrence.merge(
    total_violations,
    on="Center_Site_ID",
    how="left")

# Merge in top repeated violation
center_recurrence = center_recurrence.merge(
    top_repeat_violation,
    on="Center_Site_ID",
    how="left")

site_lookup = (
    Violations
    .groupby("Center_Site_ID")
    .agg(
        # Combine all center names at same site
        center_names=(
            "Center Name",
            lambda x: " / ".join(sorted(set(x)))),
        borough=("Borough", "first"),
        zipcode=("ZipCode", "first"),
        building=("Building", "first"))
    .reset_index())

# Merge readable center info
center_recurrence = center_recurrence.merge(
    site_lookup,
    on="Center_Site_ID",
    how="left")

center_recurrence = center_recurrence.merge(
    recurrence_by_category,
    on="Center_Site_ID",
    how="left")

recurring_category_cols = [
    "recurring_critical_violations",
    "recurring_general_violations",
    "recurring_public_health_violations",
    "recurring_blank_category_violations"]

for col in recurring_category_cols:
    if col in center_recurrence.columns:
        center_recurrence[col] = (
            center_recurrence[col]
            .fillna(0)
            .astype(int))

# Sort descending by biggest repeated single issue
center_recurrence = center_recurrence.sort_values(
    [
        "total_violations",
        "max_times_same_violation_seen",
        "repeated_violation_types",
        "recurrence_score"
    ],
    ascending=False)

center_recurrence.head(50)
```

```python
# Sort centers by recurrence score
ranked_centers = center_recurrence.sort_values(
    "recurrence_score",
    ascending=False).copy()

# Only centers with at least 1 violation
centers_with_violations = ranked_centers[
    ranked_centers["total_violations"] >= 1].copy()
def bronx_share(df, label):
    total = len(df)
    bronx = (df["borough"].str.upper() == "BRONX").sum()
    pct = bronx / total if total > 0 else 0
    return {
        "Group": label,
        "Total Centers": total,
        "Bronx Centers": bronx,
        "Bronx Share": pct}

bronx_concentration = pd.DataFrame([
    bronx_share(centers_with_violations, "All centers with violations"),
    bronx_share(centers_with_violations.head(250), "Top 250 by recurrence_score"),
    bronx_share(centers_with_violations.head(100), "Top 100 by recurrence_score"),
    bronx_share(centers_with_violations.head(50), "Top 50 by recurrence_score"),
    bronx_share(centers_with_violations.head(10), "Top 10 by recurrence_score")])

# Format percent for display
bronx_concentration["Bronx Share %"] = (
    bronx_concentration["Bronx Share"] * 100).round(1)

bronx_concentration
```
